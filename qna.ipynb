```python
import openai
from dotenv import load_dotenv
import gradio as gr
import os
```

```python

# Load environment variables from .env file
load_dotenv()

openai.api_type = os.getenv("OPENAI_API_TYPE")
openai.api_version = os.getenv("OPENAI_API_VERSION")
openai.api_base = os.getenv("OPENAI_API_BASE")
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
def generate_answer(context, question):
    prompt = f"Context: {context}\nQuestion: {question}\nAnswer:"
    completions = openai.Completion.create(
        engine="aouy-txt-davinci-003",
        prompt=prompt,
        max_tokens=4097,
        n=1,
        stop=None,
        temperature=0.1,
    )
    message = completions.choices[0].text.strip()
    return message
```

```python
iface = gr.Interface(
    fn=generate_answer,
    inputs=[gr.inputs.Textbox(label="Context"), gr.inputs.Textbox(label="Question")],
    outputs="text",
    examples=[
        ["France is a country located in Europe.", "What is the capital of France?"],
        ["Our solar system is a collection of planets, moons, asteroids, and comets.", "How many planets are in our solar system?"],
        ["The telephone is a communication device that was invented in the 19th century.", "Who invented the telephone?"],
    ],
    title="OpenAI GPT-3 Q&A",
    description="Ask any question and get an answer!",
)

iface.launch()
```
